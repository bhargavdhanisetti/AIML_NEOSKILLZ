```python
#day 13 14 15 16 17
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data=pd.read_csv("/content/Iris (1) (1).csv")
print(data.head())
```

```python
data['SepalLengthCm']=data['SepalLengthCm'].fillna(data['SepalLengthCm'].mean())
```

```python
data['SepalWidthCm']=data['SepalWidthCm'].fillna(data['SepalWidthCm'].median())
```

```python
data['PetalLengthCm']=data['PetalLengthCm'].fillna(data['PetalLengthCm'].median())
```

```python
data.isnull().sum()
```

```python
conditional_sepal_length=data[data['SepalLengthCm']<=5.5]
print(conditional_sepal_length.head())
```

```python
conditional_species=data[data['Species'] != 'Iris-setosa']
print(conditional_species)
```

```python
conditional_species_1=data[data['Species'].isin(['Iris-versicolor', 'Iris-virginica'])]
print(conditional_species_1)
```

```python
#line plot
plt.plot(data["SepalLengthCm"])
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Scatter plot
plt.scatter(data["SepalLengthCm"], data["SepalWidthCm"])
plt.title("Scatter plot")
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.show()
```

```python
plt.hist(data["PetalLengthCm"],bins=20)
plt.title("Histogram")
plt.xlabel("PetalLengthCm")
plt.ylabel("Count")
plt.show()
```

```python
plt.boxplot(data["SepalWidthCm"])
plt.title("Box plot")
plt.ylabel("Sepal Width")
plt.show()
```

```python
data["Species"].value_counts().plot(kind="pie", autopct="%0.1f%%")
plt.title("Pie chart")
plt.show()
```

```python
data.iloc[:, 1:5].plot(kind="box")
plt.title("Box plot")
plt.show()
```
